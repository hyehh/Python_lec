```python
# https://www.rottentomatoes.com/top/bestofrt/?year=2019
#top_movies_main > div > table > tbody > tr:nth-child(1) > td:nth-child(3) > a
#top_movies_main > div > table > tbody > tr:nth-child(1) > td.bold
from bs4 import BeautifulSoup
import urllib.request as req
from urllib.request import urlopen
import pandas as pd

url = "https://www.rottentomatoes.com/top/bestofrt/?year=2019"
res = req.urlopen(url)
soup = BeautifulSoup(res, "html.parser")
# print(soup)

movie_rank_list = soup.select("#top_movies_main div table td.bold")
movie_rating_list = soup.select("#top_movies_main div table span.tMeterScore")
movie_name_list = soup.select("#top_movies_main div table a")
#movie_review_list = soup.select("#top_movies_main div table .right.hidden-xs")
movie_review_list = soup.select(".table  td.right.hidden-xs")
# movie_name_list = soup.select(".table td a")

rank_lists = []
rating_lists = []
name_lists = []
review_lists = []
    
for rank in movie_rank_list:
    print(rank.string.strip())
    rank_lists.append(rank.string.strip())
    
for rating in movie_rating_list:
    print(rating.string.strip())
    rating_lists.append(rating.string.strip())
    
for name in movie_name_list:
    print(name.string.strip())
    name_lists.append(name.string.strip())
    
for review in movie_review_list:
    print(review.string.strip())
    review_lists.append(review.string.strip())
```

```python
print(rank_lists)
```

```python
print(rating_lists)
```

```python
print(name_lists)
```

```python
print(review_lists)
```

```python
df = pd.DataFrame(data=rank_lists, columns=['Rank'])
df['Rating'] = pd.Series(rating_lists)
df['Title'] = pd.Series(name_lists)
df['Review'] = pd.Series(review_lists)
df
```

```python
df.to_csv("./Data/Rotten_TomatoesLists.csv", sep=",", encoding='utf-8', index = False)
```
